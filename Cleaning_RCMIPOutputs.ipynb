### Filter Variable Imports based on FeliX's Available Set
- This step is mainly to reduce dataset size since its already massive


```python
# Read the CSV file
var_mapping_df = pd.read_csv('Vensim_Runs/FeliX_RCMIP_VarMapping.csv')
unique_variables = var_mapping_df['Variable'].unique().tolist()
print("Variables FeliX has in RCMIP\n", unique_variables)
```

### Combine the Files of Choice

```python
# Define the directory where the files are located
directory = dir_list  # Changed to use dir_list

# Helpers to standardize columns for EVERY file
def _normalize_column_name(name):
    name = str(name).strip()
    # Pure year like '1850'
    if name.isdigit():
        try:
            year = int(name)
            if 1500 <= year <= 2500:
                return str(year)
        except Exception:
            pass
    # Try parsing timestamp-like header to year
    try:
        parsed = pd.to_datetime(name, errors='raise', utc=True)
        year = parsed.year
        if 1500 <= year <= 2500:
            return str(int(year))
    except Exception:
        pass
    # Numeric like '1850.0'
    try:
        f = float(name)
        i = int(f)
        if abs(f - i) < 1e-9 and 1500 <= i <= 2500:
            return str(i)
    except Exception:
        pass
    return name


def _standardize_df(df):
    # Normalize headers (including date/year-like)
    df.columns = [_normalize_column_name(c) for c in df.columns]

    # Unify any case variants of ClimateModel/Climatemodel → ClimateModel
    variants = [c for c in df.columns if str(c).strip().lower() == 'climatemodel']
    if 'ClimateModel' not in df.columns and variants:
        df = df.rename(columns={variants[0]: 'ClimateModel'})
        variants = variants[1:]
    for v in variants:
        if v == 'ClimateModel':
            continue
        if v in df.columns:
            if 'ClimateModel' not in df.columns:
                df = df.rename(columns={v: 'ClimateModel'})
            else:
                df['ClimateModel'] = df['ClimateModel'].where(df['ClimateModel'].notna(), df[v])
                df.drop(columns=[v], inplace=True)

    # Normalize 'Variable' capitalization
    if 'Variable' not in df.columns:
        for c in list(df.columns):
            if str(c).strip().lower() == 'variable':
                df = df.rename(columns={c: 'Variable'})
                break

    return df

# Initialize an empty list to store the imported dataframes
dataframes = []
file_names = []

# Import each file automatically
for file_path in directory:
    try:
        # Get file extension
        file_extension = os.path.splitext(file_path)[1].lower()
        
        if file_extension == '.csv':
            df = pd.read_csv(file_path)
            df = _standardize_df(df)
            if 'Variable' in df.columns:
                df = df[df['Variable'].isin(unique_variables)]
            dataframes.append(df)
            file_names.append(os.path.basename(file_path))
            print(f"Successfully imported: {os.path.basename(file_path)}")
            
        elif file_extension in ['.xlsx', '.xls']:
            # For Excel files, only import the sheet called "your_data"
            df = pd.read_excel(file_path, sheet_name="your_data")
            df = _standardize_df(df)
            if 'Variable' in df.columns:
                df = df[df['Variable'].isin(unique_variables)]
            dataframes.append(df)
            file_names.append(f"{os.path.basename(file_path)} - your_data")
            print(f"Successfully imported: {os.path.basename(file_path)} - your_data")
                
        else:
            print(f"Skipping unsupported file type: {os.path.basename(file_path)}")
            
    except Exception as e:
        print(f"Error importing {os.path.basename(file_path)}: {str(e)}")

# Display summary
print(f"\nTotal files imported: {len(dataframes)}")
print(f"Files imported: {file_names}")

# Display the imported dataframes
for i, df in enumerate(dataframes):
    print(f"\nDataFrame {i+1} ({file_names[i]}):")
    print(f"Shape: {df.shape}")
    print(df.head())
    print("-" * 50)

# Append all dataframes into a combined results dataframe
RCMIP_combined = pd.concat(dataframes, ignore_index=True)

# Final pass: standardize combined headers and consolidate duplicates
RCMIP_combined.columns = [str(c).strip() for c in RCMIP_combined.columns]
RCMIP_combined.columns = [_normalize_column_name(c) for c in RCMIP_combined.columns]

if RCMIP_combined.columns.duplicated().any():
    print("Found duplicate columns after normalization, combining them...")
    unique_cols = RCMIP_combined.columns.drop_duplicates()
    combined_data = {}
    for col in unique_cols:
        duplicate_cols = [c for c in RCMIP_combined.columns if c == col]
        if len(duplicate_cols) > 1:
            combined_data[col] = RCMIP_combined[duplicate_cols].bfill(axis=1).iloc[:, 0]
            print(f"Combined duplicate columns: {duplicate_cols} -> {col}")
        else:
            combined_data[col] = RCMIP_combined[col]
    RCMIP_combined = pd.DataFrame(combined_data)

print(f"\nRCMIP_combined DataFrame Shape: {RCMIP_combined.shape}")
print(f"Columns: {list(RCMIP_combined.columns)}")
print(RCMIP_combined.head())

# Export the combined results dataframe as a csv
RCMIP_combined.to_csv('RCMIP_ModelOutputs/RCMIP_combined.csv', index=False)
print("RCMIP_combined dataframe exported as RCMIP_combined.csv")
```
